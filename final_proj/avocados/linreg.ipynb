```python
import pandas as pd
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
import statsmodels.api as sm
import numpy as np
import seaborn as sns
from scipy import stats

# preprocess data
raw = pd.read_csv('avocado.csv')

# drop useless index column
raw = raw.drop(columns=['Unnamed: 0'])

# sort by date
raw = raw.sort_values('Date')

# turn type into 0 = conventional 1 = organic
raw["type"].replace({"conventional": 0, "organic": 1}, inplace=True)
```

```python
# create the model
# drop non-numerical/non-continuous columns 
X = raw.drop(columns=[ 'region', 'AveragePrice'])

X = sm.add_constant(X)
y = raw.AveragePrice
```

```python
# check assumptions
#4046,4225,4770,Total Bags,Small Bags,Large Bags,XLarge Bags
fig, axs = plt.subplots(2, 5, figsize=(20, 10))
axs[0, 0].scatter(X[['Total Volume']], y)
axs[0, 0].set_title('AveragePrice')

axs[0, 1].scatter(X[['Total Bags']], y)
axs[0, 1].set_title('AveragePrice')

axs[0, 2].scatter(X[['4046']], y)
axs[0, 2].set_title('AveragePrice')

axs[0, 3].scatter(X[['4225']], y)
axs[0, 3].set_title('AveragePrice')

axs[1, 0].scatter(X[['4770']], y)
axs[1, 0].set_title('AveragePrice')

axs[1, 1].scatter(X[['Small Bags']], y)
axs[1, 1].set_title('AveragePrice')

axs[1, 2].scatter(X[['Large Bags']], y)
axs[1, 2].set_title('AveragePrice')

axs[1, 3].scatter(X[['Total Bags']], y)
axs[1, 3].set_title('AveragePrice')

axs[1, 4].scatter(X[['XLarge Bags']], y)
axs[1, 4].set_title('AveragePrice')

axs[0, 4].scatter(X[['year']], y)
axs[0, 4].set_title('AveragePrice')

plt.savefig('avocado_linreg_linearity_test')
#not suitable for linear regression: variables are not linear in response
```
